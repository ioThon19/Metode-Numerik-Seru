**PROGRAM SPL DARI IBU**

```python
import numpy as np
import sys

n = int(input('Input jumlah variabel: '))
a = np.zeros((n,n+1))
x = np.zeros(n)

print('Input matriks: ')
for i in range(n):
    for j in range(n+1):
        a[i][j] = float(input( 'a['+str(i)+']['+ str(j)+']='))
for i in range(n):
    if a[i][i] == 0.0:
        sys.exit('Terdapat nilai 0 pada diagonal utama!')
    for j in range(n):
        if i != j:
            ratio = a[j][i]/a[i][i]

            for k in range(n+1):
                a[j][k] = a[j][k] - ratio * a[i][k]

for i in range(n):
    x[i] = a[i][n]/a[i][i]

print('\nSolusi SPL: ')
for i in range(n):
    print('X%d = %0.2f' %(i,x[i]), end = '\t')
```

```python
import numpy as np

n = int(input('Input variabel : '))
a = np.zeros((n,n+1))
x = np.zeros(n)

print('Masukkan bilangan ke dalam matrix : ')
for i in range(n):
    for j in range(n+1):
        a[i][j] = float(input('a['+ str(i) + '][' + str(j) + '] = '))

print('-'*45)
print('----- Bagian Proses -----')

for i in range(n):
    if i != n:
        if a[i][i] == 0.0:
            search = i
            while a[i][i] == 0.0:
                for j in range(n+1):
                    temp1 = a[i][j]
                    temp2 = a[search+1][j]
                    a[i][j] = temp2
                    a[search+1][j] = temp1
                    print('a[%d][%d] = %0.2f' %(i,j,temp1))
                    print('b[%d][%d] = %0.2f' %(search+1,j,temp2))
                    print('Penukaran [%d][%d] dengan [%d][%d] berhasil!' %(i,j,search+1,j))
                    print('a[%d][%d] Sekarang = %0.2f' %(i,j,temp2))
                    print('b[%d][%d] Sekarang = %0.2f' %(search+1,j,temp1))
                search = search + 1
        
    for j in range(n):
        print('i = %d , j = %d' %(i,j))
        if i != j:
            ratio = a[j][i] / a[i][i]
            print('  ratio = a[%d][%d] / a[%d][%d]' %(j,i,i,i))
            print('  ratio = %0.2f / %0.2f = %0.2f' %(a[j][i], a[i][i], ratio))
            
            for k in range(n+1):
                print('  k = %d' %(k))
                print('  a[%d][%d] = a[%d][%d] - ratio * a[%d][%d]' %(j,k,j,k,i,k))
                print('  a[%d][%d] = %0.2f - %0.2f * %0.2f' %(j,k,a[j][k],ratio,a[i][k]))
                a[j][k] = a[j][k] - ratio * a[i][k]
                print('  a[%d][%d] = %0.2f' %(j,k,a[j][k]))

print('-'*45)
print('----- Bagian Mencari X -----')
                
for i in range(n):
    print('x%d = a[%d][%d] / a[%d][%d]' %(i+1,i,n,i,i))
    x[i] = a[i][n] / a[i][i]
    print('x%d = %0.2f'%(i+1,x[i]))

print('-'*45)

print('Solusinya : ')
for i in range(n):
    print('x%d = %0.2f' %(i+1,x[i]), end='\t')
```

**HASIL EKSPLORASI**

```python
import numpy as np

n = int(input('Input jumlah variabel: '))
a = np.zeros((n,n+1))
x = np.zeros(n)

print('Input matriks: ')
for i in range(n):
    for j in range(n+1):
        a[i][j] = float(input( 'a['+str(i)+']['+ str(j)+'] = '))
        
for i in range(n):
    # Penukar baris jika bilangan di salah satu diagonalnya = 0
    if i != n:
        if a[i][i] == 0.0:
            search = i
            while a[i][i] == 0.0:
                for j in range(n+1):
                    temp1 = a[i][j]
                    temp2 = a[search+1][j]
                    a[i][j] = temp2
                    a[search+1][j] = temp1
                search = search + 1
    
    # Proses Eliminasi GaussJordan        
    for j in range(n):
        if i != j:
            ratio = a[j][i]/a[i][i]

            for k in range(n+1):
                a[j][k] = a[j][k] - ratio * a[i][k]

# Menghitung Solusi
for i in range(n):
    x[i] = a[i][n]/a[i][i]

# Menampilkan Solusi
print('\nSolusinya adalah : ')
for i in range(n):
    print('X%d = %0.2f' %(i+1,x[i]), end = '\t')
```
